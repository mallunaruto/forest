```python
import rasterio 
import numpy as np
import matplotlib as plt
```

```python
forest_file=rasterio.open("gm_ve_v1.tif")
forests=forest_file.read()
```

```python
print(np.min(forests))
print(np.max(forests))
```

```python
print(np.unique(forests))
```

```python
print(len(forests))
```

```python
import matplotlib.pyplot as plt 
fig=plt.figure()
ax=plt.axes()
fig.set_size_inches(21,7)
imgs=plt.imshow(forests[0],cmap="Greens",interpolation='nearest')
ax.axis('off')
plt.show()
```

```python
from matplotlib.colors import ListedColormap
from matplotlib import cm
forests[0][forests[0]==254]=0.0
ourcmap=cm.get_cmap('Greens',101)
newcolors=ourcmap(np.linspace(0,1,101))

print(newcolors)
background_colour = np.array([0.9882352941176471, 0.9647058823529412, 0.9607843137254902, 1.0])
#background_colour

newcolors[:1, :] = background_colour


newcmp_forests = ListedColormap(newcolors)
```

```python
import matplotlib.pyplot as plt 
fig=plt.figure()
ax=plt.axes()
fig.set_size_inches(54,15)
imgs=plt.imshow(forests[0],cmap=newcmp_forests,interpolation='nearest')
ax.set_xlim(2000, 43500)
ax.set_ylim(19500, 800)
ax.axis('off')
plt.show()
```
